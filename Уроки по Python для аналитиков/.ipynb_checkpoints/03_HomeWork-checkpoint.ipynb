#### 1.2 Создать новый признак delta_renovated, который будет содержать разницу <br> в годах между годом реновацией дома и годом постройки дома
#### Если реновации дома не было, то в новом признаке поставьте 0


```python
def check_and_diff(row):
    if row['yr_renovated'] == 0:
        return 0
    else:
        return row['yr_renovated'] - row['yr_built']

df['delta_renovated'] = df.apply(check_and_diff, axis=1)
df[df['yr_renovated']!=0]
```

#### 1.3 Создайте признаки года продажи, месяца продажи


```python
df['year_sold'] = df['date'].apply(lambda x: x[:4])
df['month_sold'] = df['date'].apply(lambda x: x[4:6])
df.head(5)
```

```python
df_date = pd.read_csv('kc_house_data.csv', encoding = 'windows-1251', parse_dates=['date'], dayfirst=False)
df_date.head(5)
```

```python
df_date['year_sold'] = df_date['date'].dt.year
df_date['month_sold'] = df_date['date'].dt.month_name()
df_date['month'] = df_date['date'].dt.month
df_date.head(5)
```

### Условие 3: 3 задача
Составьте несколько сводных таблиц

#### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms').agg({'price': 'mean'})
```

#### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома


```python
df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
```

#### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида


```python
pd.crosstab(index = df['condition'], columns = df['waterfront'])
```

#### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
pd.crosstab(index = df['floors'], columns = df['bedrooms'])
```

#### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
pd.crosstab(index = df['condition'], columns = df['grade'], values = df['price'], aggfunc = 'median')
```
