```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Условие 1: Задача 1

#### 1.2. Изучите стоимости недвижимости

```python
df = pd.read_csv('kc_house_data.csv', encoding = 'windows-1251')
```

```python
df.dtypes
```

```python
plt.figure(figsize=(12, 3))
plt.hist(df['price'], bins=51)
plt.title('Цена в евро')
plt.xlabel('Цена')
plt.ylabel('Количество')
plt.grid()
plt.xticks(range(75000, 7700000, 150000), rotation=90);
```

##### Вывод: подавляющее большинство домов стоит в пределах 1000000 евро

#### 1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(12, 3))
plt.hist(df['sqft_living'], bins=40)
plt.title('Распределение количества жилых квадратных метов')
plt.xlabel('м2')
plt.ylabel('Количество')
plt.grid()
plt.xticks(range(290, 13540, 330), rotation=90);
```

##### Вывод: основная масса домов лежит в диапазоне от 700 м2 до 3300 м2

#### 1.4. Изучите распределение года постройки

```python
plt.figure(figsize=(12, 3))
plt.hist(df['yr_built'], bins=23)
plt.title('Распределение по годам постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество')
plt.grid()
plt.xticks(range(1900, 2015, 5), rotation=90);
```

##### Вывод: видим нарастающую тенденцию количества построенных домов с провалами в Великую Депрессию и кризиса 70-х

#### ***Построим корреляционную таблицу*** 

```python
dfcorr = df.corr(numeric_only=True).round(2)
plt.figure(figsize=(15, 8))
sns.heatmap(dfcorr, annot=True, cmap='coolwarm');
```

##### Из корреляционной таблицы видно. что наибольшее влияния на 'price' оказывают параметры **bedrooms**, **bathrooms**, **sqft_living**, **grade**, **sqft_above**, **sqft_living15**, **lat**, **view**, **sqft_basement**.
Рассмотрим некоторые из них более детально...

### **1. Рассмотрим влияние на цену параметра bedrooms**

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['bedrooms'], y=df['price'], showfliers=False)
plt.title('Зависимость цены квартиры от количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Цена');
```

#### Вывод: из графика видно, что медиана у баров в зависимости от количества спален повышательная.

### **2. Рассмотрим влияние на цену параметра bathrooms**

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['bathrooms'], y=df['price'], showfliers=False)
plt.title('Зависимость цены квартиры от количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Цена')
plt.xticks(rotation=90);
```

#### Вывод: зависимость прослеживается, но после 4.5 медиану начинает лихорадить.

### **3. Далее рассмотрим зависимость цены от размера жилой площади**

```python
sns.jointplot(y=df['sqft_living'], x=df['price'], kind='reg');
```

**Вывод: зависимость видна отчетливо**

### **4. Также рассмотрим влияние площади чердака на ценц**

```python
sns.jointplot(x=df['sqft_above'], y=df['price'], kind='reg');
```

##### **Вывод: здесь без сюрпризоы**

### **5. И наконец посмотрим влияние оценки на цену**

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['grade'], y=df['price'], showfliers=False)
plt.title('Зависимость цены квартиры от уровня оценки')
plt.xlabel('Уровень оценки')
plt.ylabel('Цена');
```

**Вывод: зависимость налицо!**
