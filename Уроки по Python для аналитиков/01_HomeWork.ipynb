### Условие 2:


На складе лежат разные фрукты в разном количестве.<br>
Нужно написать функцию, которая на вход принимает <br>любое количество названий фруктов и их количество, <br>а возвращает общее количество фруктов на складе

```python
def total_fruits(**kwargs):
    summa = 0
    for key in kwargs.keys():
        summa += kwargs[key]
    print(f'общее количество фруктов - {summa}')

total_fruits(apple = 5, kiwi = 6, banana = 15)
```

#### 4.2 Найдите выручку компании в зависимости от месяца<br>
Для этого напишите функцию, которая на вход принимает список <br>
с датами и список с выручкой, а на выходе словарь, где ключи - <br>
это месяцы, а значения - это выручка.
Используйте аннотирование типов.


```python
def summ_and_dates (data: list, summ: list) -> dict:
    res = {}
    for i in range(1, 13):
        if i < 10:
            month = f'-0{str(i)}-'
        else:
            month = f'-{str(i)}-'
        total_sum = sum([summ[i] for i in range(len(dates)) if month in data[i]])
        res[i] = total_sum
    return res

for k, v in summ_and_dates(dates, summes).items():
    print(f'в {k} месяц общая выручка {v}')
```
